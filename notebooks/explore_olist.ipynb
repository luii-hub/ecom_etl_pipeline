```python
# Import requirements
import pandas as pd
import numpy as np
```

```python
customers_df = pd.read_csv('../data/olist_customers_dataset.csv')
geolocation_df = pd.read_csv('../data/olist_geolocation_dataset.csv')
order_items_df = pd.read_csv('../data/olist_order_items_dataset.csv')
order_payments_df = pd.read_csv('../data/olist_order_payments_dataset.csv')
order_reviews_df = pd.read_csv('../data/olist_order_reviews_dataset.csv')
orders_df = pd.read_csv('../data/olist_orders_dataset.csv')
products_df = pd.read_csv('../data/olist_products_dataset.csv')
sellers_df = pd.read_csv('../data/olist_sellers_dataset.csv')
product_category_name_translation_df = pd.read_csv('../data/product_category_name_translation.csv')
```

```python
dataframes = {
    "Customers": customers_df,
    "Geolocation": geolocation_df,
    "Order Items": order_items_df,
    "Order Payments": order_payments_df,
    "Order Reviews": order_reviews_df,
    "Orders": orders_df,
    "Products": products_df,
    "Sellers": sellers_df,
    "Product Category Name Translation": product_category_name_translation_df
}   
```

```python
# Display the first few rows of each dataframe
for name, df in dataframes.items():
    print(f"\n{name} DataFrame:")
    print(df.head())
```

```python
# Check the information of each dataframe
for name, df in dataframes.items():
    print(f"\n{name} DataFrame:")
    print(df.info())
```

```python
# Check for missing values in each dataframe
for name, df in dataframes.items():
    print(f"\n{name} DataFrame Missing Values:")
    print(df.isnull().sum())
```

```python
# Check for blank strings in columns
for name, df in dataframes.items():
    print(f"\n{name} DataFrame Blank String Counts:")
    blank_counts = (df == '').sum()
    print(blank_counts[blank_counts > 0])
```

```python
# Check for duplicates in each dataframe
for name, df in dataframes.items():
    print(f"\n{name} DataFrame:")
    print(df.duplicated().sum())
```
